### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
# ВАШ КОД ЗДЕСЬ
arr = np.arange(10,50)
print(f"Исходный массив:\n {arr} ")

arr = arr[::-1]
print(f"Развернутый массив:\n {arr} ")

arr = arr[arr % 2 ==0]
print(f"Четные элементы:\n {arr} ")
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
# ВАШ КОД ЗДЕСЬ
arr = np.random.rand(4,3,2)
print(f"Исходный массив:\n {arr} \n")
print(f"Минимальный элемент массива: {np.min(arr)} \n")
print(f"Максимальный элемент массива: {np.max(arr)} ")
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
# ВАШ КОД ЗДЕСЬ
np.random.seed(42)
arr4х3 = np.random.rand(4,3)
print(f"Массив 4x3:\n {arr4х3} \n")

np.random.seed(42)
arr6х4 = np.random.rand(6,4)
print(f"Массив 6х4:\n {arr4х3} \n")

result = arr6х4 @ arr4х3
print(f"Результат умножения массивов:\n {arr4х3} ")
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
# ВАШ КОД ЗДЕСЬ
np.random.seed(42)
arr = np.random.rand(7,7)
print(f"Исходный массив:\n {arr} \n")

mean = np.mean(arr)
print(f"Среднее отклонение: {mean} \n")

std = np.std(arr)
print(f"Стандартное отклонение: {std} \n")

arr = (arr - mean) / std
print(f"Нормализованный массив:\n {arr} ")
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
# ВАШ КОД ЗДЕСЬ
import pandas as pd
import matplotlib as mpl
import seaborn as sns
```

### Загрузите датасет Tips из набора датасетов seaborn

```python
# ВАШ КОД ЗДЕСЬ
tips = sns.load_dataset('tips')
```

### Посмотрите на первые 5 строчек

```python
# ВАШ КОД ЗДЕСЬ
tips.head(5)
```

### Узнайте сколько всего строчек и колонок в данных

```python
# ВАШ КОД ЗДЕСЬ
print(f"Количество строк в данных: {tips.shape[0]}")
print(f"Количество колонок в данных: {tips.shape[1]}")
```

### Проверьте есть ли пропуски в данных

```python
# ВАШ КОД ЗДЕСЬ
tips.isna().any()
```

### Посмотрите на распределение числовых признаков

```python
# ВАШ КОД ЗДЕСЬ
tips['total_bill'].value_counts()
```

### Найдите максимальное значение 'total_bill'

```python
# ВАШ КОД ЗДЕСЬ
max = tips['total_bill'].max()
print(f"Mаксимальное значение 'total_bill': {max}")
```

### Найдите количество курящих людей

```python
# ВАШ КОД ЗДЕСЬ
smokers = tips['smoker'].where(tips['smoker'] == 'Yes').count()
print(f"Количество курящих: {smokers}")
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
# ВАШ КОД ЗДЕСЬ
result = tips.groupby('day')['total_bill'].mean().reset_index()
print(result)
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
# ВАШ КОД ЗДЕСЬ
median = (tips['total_bill'].median())
print(f'Медиана для total_bill: {median}')

df = tips[(tips['total_bill'] > median)]

result = df.groupby('sex')['tip'].mean().reset_index()
print(result)
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
# ВАШ КОД ЗДЕСЬ

m = {"No" : 0,"Yes" : 1}
tips['smoker'] = tips['smoker'].map(m).astype('int64')
tips
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
# ВАШ КОД ЗДЕСЬ

tips['total_bill'].plot(
    kind='hist',
    bins=50,
    title="Распределение счетов"
)
plt.xlabel('среднй счет')
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
# ВАШ КОД ЗДЕСЬ
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip'
)
plt.title("Чаевые от счета")
```

### Постройте pairplot

```python
# ВАШ КОД ЗДЕСЬ
sns.pairplot(tips)
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
# ВАШ КОД ЗДЕСЬ
sns.violinplot(
    data=tips,
    x='day',
    y='total_bill',
    split=True
)
plt.title('Распределение счетов по дням недели');
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
# ВАШ КОД ЗДЕСЬ
sns.histplot(
    data=tips,
    x='tip',
    hue='time'
)
plt.xlabel('tip');
plt.title('Распределение tip от time');
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
# ВАШ КОД ЗДЕСЬ

male = tips[tips['sex'] == 'Male']
female = tips[tips['sex'] == 'Female']

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.scatterplot(
    data=male,
    x='total_bill',
    y='tip',
    hue='smoker',
    ax=axes[0]
)
axes[0].set_title('Male')

sns.scatterplot(
    data=female,
    x='total_bill',
    y='tip',
    hue='smoker',
    ax=axes[1]
)
axes[1].set_title('Female')

plt.show()
```
